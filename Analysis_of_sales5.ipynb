```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
vgms = pd.read_csv('/content/vgsales.csv')
display(vgms)
```

```python
plt.figure(figsize=(10,5))
plt.scatter( vgms['Year'],vgms['Global_Sales'], s = 15)
plt.xlabel('Year')
plt.ylabel('Global_Sales')
plt.title('video games sales')
plt.show();
```

the global sales increased in 2010

```python
columns = [vgms['NA_Sales'].sum(), vgms['EU_Sales'].sum(), vgms['JP_Sales'].sum(), vgms['Other_Sales'].sum()]
labels = ['NA_Sales', 'Europe', 'Japan', 'Other_Sales']
explode = [0.05, 0.05, 0.05, 0.05]

plt.rc('figure', figsize = (6,6))

fig, ax = plt.subplots()
ax.pie(columns, labels = labels, autopct = '%.1f%%',  explode = explode)

centre_circle = plt.Circle((0, 0), 0.7, fc='white')
ax.add_artist(centre_circle)

plt.show()
```

the NA_sales the most sales
